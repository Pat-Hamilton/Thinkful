```python
import numpy as np
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

mpg = sns.load_dataset('mpg')
```

```python
ax = sns.stripplot(x=mpg['mpg'])
```

- Pro: Can see each point
- Con: Harder time seeing the spread and median

```python
ax = sns.boxplot(x=mpg['mpg'])
```

- Pro: Can see a rough distribution of MPG
- Con: Doesn't show individual values

```python
ax = sns.countplot(x=mpg['mpg'])
```

- Pro: Can see the distribution in a little more detail than a boxplot 
- Con: Harder time finding the median

```python
ax = sns.violinplot(x=mpg['mpg'])
```

- Pro: Can see probability density
- Con: Not as widely known

```python
bx = sns.boxplot(x = 'cylinders', y = 'mpg', data = mpg)
```

- Pro: Can see a rough distribution of MPG for each cylinder amount
- Con: Doesn't show individual values

```python
vl = sns.violinplot(x = 'cylinders', y = 'mpg', data = mpg)
```

- Pro: Can see probability density
- Con: Not as widely known

```python
ct = sns.catplot(x = 'cylinders', y = 'mpg', data = mpg)
```

- Pro: Can see how many datapoints there are for each cylinder count
- Con: No clear median mark

```python
ct = sns.pointplot(x = 'cylinders', y = 'mpg', data = mpg)
```

- Pro:Can see both trend and error bars
- Con: Can't see how many datapoints there are

```python
ln = sns.lineplot(x = 'cylinders', y = 'mpg', data = mpg)
```

- Pro: Good at showing trends
- Con: Better for continuous data

```python
br = sns.barplot(x = 'cylinders', y = 'mpg', data = mpg)
```

- Pro: Clearly show the differnce in mean MPG between number of cylinders
- Con: Inefficient use of space

```python
ln2 = sns.lineplot(x = 'weight', y = 'acceleration', data = mpg)
```

- Pro: Can see a general downward trend pretty clearly
- Con: It looks a little crowded

```python
sc = sns.scatterplot(x = 'weight', y = 'acceleration', data = mpg)
```

- Pro: Can see a downward trend and individual records
- Con: Correlation might not always be clear. Might be too crowded if too many datapoints

```python
jt = sns.jointplot(x = 'weight', y = 'acceleration', data = mpg, kind = 'regression')
```

- Pros: Can see the density of each variable's values isolated from the other variable
- Cons: Might be too much going on if you don't care about each variables density. Might be best left in the exploratory stage

```python
sns.set(style="ticks")

sns.pairplot(mpg, vars = ['mpg', 'displacement', 'horsepower', 'weight'], hue="origin")
```

- Horsepower, displacement, and weight all have a positive correlation with each other, while mpg has a negative correlation with the other three.

- American vehicles tend to be be heavier with more displacement and horsepower and a lower mpg, while European and Japanese tend to be the opposite.
