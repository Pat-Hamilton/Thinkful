```python
import numpy as np
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv(r'C:\Users\Pat\Documents\Thinkful\Files\WELLCOME_APCspend2013_forThinkful_UTF8.csv')
df=df.rename(columns = {'COST (£) charged to Wellcome (inc VAT when charged)':'Cost'})

# Get rid of $/£
df['Cost'] = df['Cost'].str.replace('$', '')
df['Cost'] = df['Cost'].str.replace('£', '')

# Covnert to float
df['Cost'] = df['Cost'].astype(float)

# Strip spaces at beginning/end
df['Journal title'] = df['Journal title'].str.lower()
df['Journal title'] = df['Journal title'].str.strip()

# Clean data
df['Journal title'] = df['Journal title'].str.replace('plos 1', 'plos one')
df['Journal title'] = df['Journal title'].str.replace('plos  one', 'plos one')
df['Journal title'] = df['Journal title'].str.replace('plosone', 'plos one')
df['Journal title'] = df['Journal title'].str.replace('neurolmage', 'neuroimage')
df['Journal title'] = df['Journal title'].str.replace('nucleic acid research', 'nucleic acids research')

# Removes 0 value
indexNames = df[ df['Cost'] == 0 ].index
df.drop(indexNames , inplace=True)

# Filter out outliers. Use catplot below to get the gist of where they are
# ct = sns.catplot(x = 'Publisher', y = 'Cost', data = df)
df = df[df['Cost'] < 12000]


# Alternative sort/count method
# df_cnt = df['Journal title'].groupby(df['Journal title']).count().sort_values(ascending=False)

print('Top 5 Jounrals by Article Count')
print(df['Journal title'].value_counts().head(n=5))

top5 = df['Journal title'].value_counts().head(n=5).index.tolist()

df2 = df[df['Journal title'].isin(top5)]
df2.groupby('Journal title').describe()



```
