```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import gzip

with gzip.open(r'C:\Users\Pat\Documents\Thinkful\Files\title.ratings.tsv.gz', 'rt') as f:
    df = pd.read_csv(f, nrows=10000, sep='\t', engine='python')
plt.hist(df['averageRating'])
plt.show()

# Nothing surprising, you should expect most movies to be somewhere around average. Also not surprised to see it's skewed left, as there are definitely more bad movies out there than good.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import gzip

with gzip.open(r'C:\Users\Pat\Documents\Thinkful\Files\title.ratings.tsv.gz', 'rt') as f:
    df = pd.read_csv(f, nrows=10000, sep='\t', engine='python')
plt.scatter(x=df['numVotes'], y=df['averageRating'])
plt.show()

# This seems pretty intuitive to me. Movies with less votes are more likely to have a more extreme rating by the law of large numbers.
```
